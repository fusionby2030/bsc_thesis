```python

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
SMALL_SIZE = 32
MEDIUM_SIZE = 35
BIGGER_SIZE = 26

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
df = pd.read_csv('/u/76/kita1/unix/Desktop/filtered_dataset.csv')
input_columns = ['averagetriangularity', 'a(m)', 'Ip(MA)', 
                 'plasmavolume(m3)', 'q95', 'gasflowrateofmainspecies1022(es)', 'B(T)', 
                 'P_ICRH(MW)', 'P_NBI(MW)', 'P_TOTPNBIPohmPICRHPshi(MW)', 
                 'H(HD)', 'Meff']
formated_names = ['$\delta$', '$a$', '$I_P$', '$V_P$', '$q_{95}$', '$\Gamma$', '$B_T$', '$P_{ICRH}$', '$P_{NBI}$', '$P_{TOT}$', '$H$', '$M_{eff}$']
input_df = df[input_columns]
```

```python
f = plt.figure(figsize=(19, 15))
plt.matshow(input_df.corr(), fignum=f.number, cmap='cividis')
plt.xticks(range(len(formated_names)), formated_names, weight='bold')
plt.yticks(range(len(formated_names)), formated_names)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=28)
plt.show()
```
